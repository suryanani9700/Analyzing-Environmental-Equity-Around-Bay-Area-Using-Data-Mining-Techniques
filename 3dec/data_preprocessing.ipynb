```python
import pandas as pd
```

```python
# Load the data
calenviroscreen_data_dictionary_path = 'calenviroscreen40resultsdatadictionary_f_2021.csv'
calenviroscreen_data_dictionary = pd.read_csv(calenviroscreen_data_dictionary_path)

calenviroscreen_data_dictionary.head()
```

```python
cal_env_df = calenviroscreen_data_dictionary[['Census Tract','Total Population', 'California County', 'ZIP','Approximate Location', 'Longitude', 'Latitude', 'Ozone','PM2.5', 'Diesel PM', 'Drinking Water', 'Lead', 'Pesticides','Tox. Release', 'Traffic', 'Cleanup Sites','Groundwater Threats', 'Haz. Waste', 'Imp. Water Bodies','Solid Waste', 'Pollution Burden', 'Asthma','Low Birth Weight','Cardiovascular Disease', 'Education', 'Linguistic Isolation','Poverty', 'Unemployment','Housing Burden']]
cal_env_df.head()
```

```python
# Load the demographic profile data
demographic_profile_path = 'demographic_profile.csv'
demographic_profile = pd.read_csv(demographic_profile_path)

print(demographic_profile.head())
```

```python
# Merging the population data columns with demographic profile data
merged_df = pd.merge( left= cal_env_df, right= demographic_profile[['Census Tract','Children < 10 years (%)','Pop 10-64 years (%)','Elderly > 64 years (%)']],left_on= 'Census Tract',right_on='Census Tract')
merged_df.columns
```

```python
merged_df['California County'].unique()
```

```python
# Selecting Bay Area data
bay_area_counties = ['Alameda ', 'Contra Costa', 'Marin ', 'Napa ', 'San Francisco','San Mateo', 'Santa Clara', 'Solano ']
ces_bay_df = merged_df[merged_df['California County'].isin(bay_area_counties)]
ces_bay_df['California County'].unique()
```

```python
# Check the data types and look for missing values
ces_bay_df.info()
```

```python
na_cols = []
type(ces_bay_df.columns)
for x in ces_bay_df.columns:
    if (ces_bay_df[x].isna().sum() != 0):
        na_cols.append(x)
    print (x , ces_bay_df[x].isna().sum())

ces_bay_df.to_csv('ces_bay.csv', index=False)
```

```python
print(na_cols)

for x in na_cols:
    ces_bay_df[x] = ces_bay_df.groupby('California County')[x].transform(lambda x: x.fillna(x.median()))

# Fill missing values with median from same county to preserve regional specificity
```
